```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##histogramas
```

```python
x=np.random.normal(size=100)
```

```python
plt.hist(x)
```

```python
plt.hist([1,2,3,4])
```

```python
# el parametro beans, nos indica la cantidad de contenedores que tendrá nuestro histograma, puede ser un entero, una secuencia o un string
#bean hace referencia a un contenedor, un espacio
```

```python
#aqui le decimos que el contenedor sea desde -5 hasta 5 y tenga 20 contenedores
plt.hist(x,bins=np.linspace(-5,5,20))
```

```python
# los subplots es una caja contenedora de plots, tiene un orden al igual que las matrices y las graficas se acomodarán comenzando de 1, incrementando primero las filas,
#luego las columnas
```

```python
#generaremos una distribucion gamma
x=np.random.gamma(1,2,1000)
```

```python
plt.hist(x,bins=20)
```

```python
#cumulative, para hacer un histograma acumulativo, si ponemos en -1 la acumulacion se invierte
plt.hist(x,cumulative=True)
```

```python
plt.hist(x,normed=True,cumulative=True,histtype="step")
```

```python
plt.subplot(333)
plt.hist(x,normed=True,cumulative=True,histtype="step")
plt.subplot(3,3,2)
plt.hist(x,cumulative=True)
plt.subplot(331)
plt.hist(x,bins=20)
```
