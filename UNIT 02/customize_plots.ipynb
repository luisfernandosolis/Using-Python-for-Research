```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(1,10,20)
y=x**2
```

```python
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.plot(x,y,"bo-",linewidth=2,markersize=7)
```

```python
# AJUSTANDO LOS EJES DEL GRAFICO
#plt.axis([xmin,xmax,ymin,ymax])
#xmin: indica donde comienza el eje x
#xmax: indica donde termina el eje x
#ymin: indica donde inicia el eje y
#ymax: indica donde termina el eje y
plt.axis([-1.5,10,-20,110])
plt.plot(x,y)
```

```python
#para definir leyendas, primero definamos las etiquetas de las graficas
#con el parametro loc: indicamos donde se graficar√° la leyenda.
x2=np.linspace(1,30,40)
y2=x2

plt.plot(x2,y2,label="second graphic")
plt.plot(x,y,label="first graphic")
plt.legend(loc="upper left")
```

```python
# para guardar la figura usamos
plt.plot(x2,y2,label="second graphic")
plt.plot(x,y,label="first graphic")
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.legend(loc="upper left")
plt.savefig("miPic.pdf")
```
